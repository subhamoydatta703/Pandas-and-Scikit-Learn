### DataFrame vs Series
#### DataFrame: Full 2D table (Excel spreadsheet)
#### Series: Single column (One column from that spreadsheet)

```python
# Creating a DataFrame
data = {
    'Name':  ['Alice', 'Bob', 'Carol', 'Dave'],
    'Score': [85, 92, 78, 95],
    'Grade': ['B', 'A', 'C', 'A']
}

# print(data)

df = pd.DataFrame(data)
print(df)
```

```python
# Creating a series
s = df['Score'] #single column of the data frame
print(s)
print(type(s)) 
```

#### .loc[]  -> by LABEL (name of the index)
##### df.loc[0]               # row with index label 0
##### df.loc[0, 'Name']       # row 0, column 'Name' -> 'Alice'
##### df.loc[0:2]             # rows 0, 1, 2 (INCLUSIVE)

#### .iloc[] -> by POSITION (like a Python list)
##### df.iloc[0]              # first row
##### df.iloc[0, 1]           # row 0, column position 1 -> 85
##### df.iloc[0:2]            # rows 0, 1 only (EXCLUSIVE)

```python
import numpy as np
thisData = {
    'Name':  ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'Score': [85, np.nan, 78, 95, np.nan],
    'Grade': ['B', 'A', None, 'A', 'B']
}

newData = pd.DataFrame(thisData)
print(newData)
```

```python
# newData.isna().sum()          # count missing values per column
# newData.dropna()              # remove rows with ANY missing value
# newData.dropna(subset=['Score'])  # remove only if Score is missing

# newData.fillna(0)                          # Fill ALL missing with 0
newData['Grade'] = newData['Grade'].fillna('NA')
newData['Score'] = newData['Score'].fillna(newData['Score'].mean()) # fill with mean (common in ML)
```

```python
print(newData)
```
